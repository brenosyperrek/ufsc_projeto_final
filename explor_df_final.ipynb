```python
#select algumas colunas pra anlisar a evolução temporal
time = df_final.groupby(pd.Grouper(key='DATA_HORA_BOLETIM', freq='M')).count()
```

```python
time.columns
```

```python
time.drop(columns=['Unnamed: 0', 'DATA_INCLUSAO','COD_REGIONAL', 'DESC_REGIONAL','ORIGEM_BOLETIM','VALOR_UPS', 'DESC_UPS', 'NUMERO_MUNICIPIO', 
                   'NOME_MUNICIPIO', 'SEQUENCIA_LOGRADOUROS', 'NUMERO_LOGRADOURO','NOME_LOGRADOURO', 'NUMERO_BAIRRO', 
                   'NOME_BAIRRO', 'NUMERO_IMOVEL', 'DESC_TIPO_ACIDENTE','DESC_TEMPO', 'PAVIMENTO',
                  'DESCRICAO_TIPO_BAIRRO'], inplace = True)
```

```python
time.info()
```

```python
time.reset_index(inplace = True)
time.rename(columns={'NUMERO_BOLETIM' : 'OCORRENCIAS'}, inplace = True)
time['AUMENTO'] = time['OCORRENCIAS'].diff()
time['ACELERACAO'] = time['AUMENTO'].diff()
time.head()
```

```python
#algum insight temporal?
def plot_time(x, y1, y2, y3, dataset, titulo):
  plt.figure(figsize=(16,12))
  ax = plt.subplot(3,1,1)
  ax.set_title(titulo, fontsize=18, loc='left')
  sns.lineplot(x=x, y=y1, data=time)
  plt.subplot(3,1,2)
  sns.lineplot(x=x, y=y2, data=time)
  plt.subplot(3,1,3)
  sns.lineplot(x=x, y=y3, data=time)
  ax=ax
```

```python
plot_time('DATA_HORA_BOLETIM','OCORRENCIAS', 'AUMENTO', 'ACELERACAO', time, 'Análise da evolução das ocorrência por ano')
```

```python
#alguma correlação?
cor = df_final.drop(columns=['Unnamed: 0', 'NUMERO_BOLETIM', 'DATA_HORA_BOLETIM', 'DATA_INCLUSAO',
      'DESC_TIPO_ACIDENTE', 'DESC_TEMPO','PAVIMENTO', 'COD_REGIONAL', 'DESC_REGIONAL',
       'ORIGEM_BOLETIM', 'VALOR_UPS', 'DESC_UPS', 'NUMERO_MUNICIPIO',
       'NOME_MUNICIPIO', 'SEQUENCIA_LOGRADOUROS', 'NUMERO_LOGRADOURO',
       'NOME_LOGRADOURO', 'NUMERO_BAIRRO', 'NOME_BAIRRO',
       'TIPO_BAIRRO', 'DESCRICAO_TIPO_BAIRRO', 'NUMERO_IMOVEL'], inplace = False)
```

```python
cor.info()
```

```python
#alguma correlação?
corr_pearson = cor.corr() #default é person
sns.heatmap(corr_pearson, cmap="Spectral", annot=True)
```

```python
corr_spear = cor.corr('spearman') #default é person
sns.heatmap(corr_spear, cmap="Spectral", annot=True)
```
